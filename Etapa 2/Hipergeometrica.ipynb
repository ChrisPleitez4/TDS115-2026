```python
# 1. importar librerias
import math as m
import matplotlib.pyplot as plt
#2. Definir los parametros del experimento
N= 108 
K=8 
n=7
x_especifico = 2
#3. Funciones para calcular la combinacion y la  distribuccion hipergeometrica
def combinacion (n,r):
    d = n-r
    C = m.factorial(n)/(m.factorial(r) *m.factorial(d))
    return C

def hipergeometrica (N,K,n,x):
    a = combinacion(K,x)
    b = combinacion(N-K,n-x)
    c = combinacion(N,n)
    H = (a*b)/c
    return H

# 3.Calcular e imprimir todas las Probbilidades de sacar 1,2... 7 comodines en una mano de 7 cartas
valores_x = []
probabilidades = []

print("Distribución teórica de comodines en una mano de 7 cartas:")
for x in range(0, min(K,n)+1):
    prob = hipergeometrica(N,K,n,x)*100
    print(f"P(X = {x}) = {prob:.8f} %")
    valores_x.append(x)
    probabilidades.append(prob)

```

```python
# 4--- Gráfica ---
plt.bar(valores_x, probabilidades, color='skyblue', edgecolor='black')
plt.axvline(x_especifico, color='red', linestyle='dashed', label=f'X = {x_especifico}')
plt.title("Distribución teórica de comodines en una mano de 7 cartas de UNO")
plt.xlabel("Número de comodines")
plt.ylabel("Probabilidad(%)")
plt.xticks(valores_x)
plt.legend()
plt.show()
```
